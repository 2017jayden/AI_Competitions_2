```python
import numpy as np
import pandas as pd 
```

```python
train_label = pd.read_csv('../input/open-images-2019-visual-relationship-truth/challenge-2019-train-vrd-labels.csv')
train = pd.read_csv('../input/open-images-2019-visual-relationship-truth/challenge-2019-train-vrd.csv')
```

```python
df = train.groupby(['LabelName1', 'RelationshipLabel', 'LabelName2']).size().reset_index(name='N')
df
```

```python
dl = df.sort_values(by=['N', 'LabelName1'], ascending=[False, True]).reset_index(drop=True)
dl
```

```python
nd = train.loc[(train['LabelName1'] == '/m/04bcr3') & (train['RelationshipLabel'] == 'is') & (train['LabelName2'] == '/m/083vt')]
```

```python
ndm = nd.median()
```

```python
ndm
```

```python
df2 = pd.DataFrame(index=range(0,329), columns=list(ndm.index))
```

```python
for i in range(len(dl)):
    row = dl.iloc[i]
    group = train.loc[(train['LabelName1'] == row[0]) & (train['RelationshipLabel'] == row[1]) & (train['LabelName2'] == row[2])]
    med = group.median()
    df2.iloc[i] = med.values
```

```python
df2
```

```python
df3 = pd.concat([dl,df2],axis=1)
df3
```

```python
df4 = df3.iloc[:300]
df4
```

```python
df5 = pd.DataFrame(index=range(0,300), columns=['pred'])
```

```python
nsum = df4['N'].sum()
```

```python
for i in range(len(df5)):
    row = df4.iloc[i]
    pred =  str(row['N']/nsum) + ' ' + \
            row['LabelName1'] + ' ' + \
            str(row['XMin1']) + ' ' + \
            str(row['YMin1']) + ' ' + \
            str(row['XMax1']) + ' ' + \
            str(row['YMax1']) + ' ' + \
            row['LabelName2'] + ' ' + \
            str(row['XMin2']) + ' ' + \
            str(row['YMin2']) + ' ' + \
            str(row['XMax2']) + ' ' + \
            str(row['YMax2']) + ' ' + \
            row['RelationshipLabel']
    df5.iloc[i] = pred
```

```python
df5
```

```python
sub = pd.read_csv('../input/open-images-2019-visual-relationship/VRD_sample_submission.csv')
```

```python
madness = ''
for i in range(100):
    madness += df5.iloc[i]['pred']
    madness += ' '
madness = madness[:-1]
```

```python
for i in range(len(sub)):
    sub.iloc[i]['PredictionString'] = madness
```

```python
sub.to_csv('submission100.csv', index=False)
```
